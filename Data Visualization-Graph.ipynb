# Creating Multiplot on Same Canvas

```python
#plt.subplot(nrows,ncols,plot_number)
#plt.tight_layout can be used to free some spaces in-between plots
fig=plt.figure()
plt.subplot(151)
plt.tight_layout()
plt.plot(x,y,"r--")

plt.subplot(1,5,2)
plt.tight_layout()
plt.plot(y,x,"g^-")

plt.subplot(1,5,3)
plt.tight_layout()
plt.plot(x,z,"y*-")

plt.subplot(1,5,4)
plt.tight_layout()
plt.plot(z,y,"c-")

plt.subplot(1,5,5)
plt.tight_layout()
plt.plot(x,y,"o--")

fig.savefig("C:\\Users\\LENOVO\Downloads\\Subplot.png")
```

```python
plt.subplot(6,1,1)
plt.tight_layout()
plt.plot(x,y,"r--")

plt.subplot(6,1,2)
plt.tight_layout()
plt.plot(y,x,"g^-")

plt.subplot(6,1,3)
plt.tight_layout()
plt.plot(x,z,"y*-")

plt.subplot(6,1,4)
plt.tight_layout()
plt.plot(z,y,"c-")

plt.subplot(6,1,5)
plt.tight_layout()
plt.plot(z,y,"gold")

plt.subplot(6,1,6)
plt.tight_layout()
plt.plot(z,x,"orange")
```

# Matplotlib Object Oriented Method
INCREASING A PLOT SIZE CAN BE DONE VIA 2 METHODS

# METHOD 1

```python

#Create figure(empty canvas)
fig=plt.figure()

#add set of axes to figure
axes=fig.add_axes([2.1,2.2,1.8,1.9])#left,bottom,width,height

#plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
plt.show()
```

# METHOD 2

```python
k=plt.figure(figsize=(15,15))
plt.plot(x,y,'b')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
fig=plt.figure()

#Create blank canvas
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom,width,height
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#inset axes

#plot on that set of axes
axes1.plot(x,y,'b')
axes1.set_xlabel("X-Axis1")
axes1.set_ylabel("Y-Axis1")
axes1.set_title("Axes 1")

axes2.plot(y,x,'r')
axes2.set_xlabel("X-Axis2")
axes2.set_ylabel("Y-Axis2")
axes2.set_title("Axes 2");

fig.savefig("C:\\Users\\LENOVO\Downloads\\Plot_within_Plot.jpeg")
```

```python
import numpy as np
plt.subplot()
xx=np.linspace(-10,10,150)
cos=np.cos(xx)
sin=np.sin(xx)

plt.plot(xx,cos,color="red",linewidth=2.4);
plt.show()
plt.subplot(212)

plt.plot(xx,sin,color="blue",linewidth=3.4);
```
