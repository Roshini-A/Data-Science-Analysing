# Categorical Data Plot

2.boxplot

3.violinplot

4.barplot

5.stripplot

6.swarmplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stores=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\Stores.csv")
tips=sns.load_dataset("tips")
```

```python
stores.head()
```

# barplot

```python

sns.barplot(x='Location',y='TotalSales',data=stores);
```

```python
sns.barplot("sex","total_bill",data=tips)
```

```python
sns.barplot(x='smoker',y='total_bill',data=tips,estimator=np.sum);
```

# countplot
this is same as barplot except it explicitly counting number of occurences,which is why we only pass x value

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
```

# boxplot

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette='coolwarm')
```

```python
sns.boxplot(x="Location",y="TotalSales",data=stores,palette='husl')
```

```python
#can do the entire datafram with orient='-h'
sns.boxplot(data=tips,palette='rainbow',orient='-h')
```

# violinplot

```python
sns.violinplot(x="day",y="total_bill",data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue='sex',palette='Set1')
```

# stripplot
  stripplot will draw scatterplot where one variable is categorical.A strip plot good on its own,but it is also a good component to box or violon plot where you want to show all observation along with underlying distribution

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

# swarmplot
swarmplot is same as stripplot but the points are adjusted such that they don't overlap.This gives a better representation of
distribution of values,although it does not scale as well to large obsevations

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

# Combining Categorical Data

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',split=True)
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

# factorplot
it is most general categorical plot.It can take in kind parameter to adjust the plot type:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="bar")
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="violin")
#kind:strip,bar,box,violin,swarm,boxen
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="swarm")
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="strip")
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="boxen")
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind="box")
```

# PieChart

```python
cars=['Audi','BMW','Ford','Tesla','Jaguar','Mercedes']
data=[21,17,35,19,12,14]
plt.pie(data,labels=cars)
plt.show()
```
